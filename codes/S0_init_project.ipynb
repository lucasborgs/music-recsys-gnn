```python
# Inicialização do projeto (criação das pastas e config.yaml)
import os
from pathlib import Path
import yaml

# Estrutura de caminhos
paths = {
    "raw": "TCC/data/raw",
    "interim": "TCC/data/interim",
    "processed": "TCC/data/processed",
    "graphs_bipartite": "TCC/graphs/bipartite",
    "graphs_item": "TCC/graphs/item_item",
    "graphs_coarse": "TCC/graphs/coarse",
    "metrics": "TCC/reports/metrics"
}

# Hiperparâmetros padrão
params = {
    "half_life_days": 180,
    "alpha": 0.5,
    "beta": 0.3,
    "gamma": 0.2,
    "k_knn": 50,
    "louvain_resolution": 1.0,
    "ppr_alpha": 0.15,
    "seed": 42
}

# Cria todas as pastas
for key, path in paths.items():
    Path(path).mkdir(parents=True, exist_ok=True)

# Pasta para configs
Path("conf").mkdir(parents=True, exist_ok=True)

# Salva config.yaml
config = {"paths": paths, "params": params}
with open("conf/config.yaml", "w") as f:
    yaml.safe_dump(config, f, sort_keys=False, allow_unicode=True)

print("Estrutura de pastas criada e conf/config.yaml gerado.")
```
